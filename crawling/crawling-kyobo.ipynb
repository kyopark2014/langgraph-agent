```python
def get_product_list(keyword: str) -> list:
    """Search product list by keyword and then return product list"""
    url = f"https://search.kyobobook.co.kr/search?keyword={keyword}&gbCode=TOT&target=total"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        prod_info = soup.find_all("a", attrs={"class": "prod_info"})
        prod_list = [
            {"title": prod.text.strip(), "link": prod.get("href")} for prod in prod_info
        ]
        return prod_list
    else:
        return []
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ChromeOptions

options = ChromeOptions()
# options.add_argument("--headless=new")
driver = webdriver.Chrome(options=options)

keyword = "텐서플로"
url = f"https://search.kyobobook.co.kr/search?keyword={keyword}&gbCode=TOT&target=total"
driver.get(url)
```

```python
prod_item = driver.find_elements(By.CLASS_NAME, "prod_info")
```

```python
for prod in prod_item[:10]:
    print(prod.text.strip())
    print(prod.get_attribute("href"))
```

```python
driver.get("https://product.kyobobook.co.kr/detail/S000000833270")
table_of_contents = driver.find_element(By.CLASS_NAME, "book_contents_item").text
print(table_of_contents)
```

```python
def get_prod_table_of_content(url: str) -> str:
    """Get product content by url"""
    options = ChromeOptions()
    options.add_argument("--headless=new")
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    contents = driver.find_element(By.CLASS_NAME, "book_contents_item").text
    driver.quit()
    return contents
```

```python
table_of_contents = get_prod_table_of_content(
    "https://product.kyobobook.co.kr/detail/S000001809989"
)
print(table_of_contents)
```
